```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
mu = 0      # mean
sigma = 1   # standard deviation

# x-axis: from -4 sigma to +4 sigma, in sigma units
x = np.linspace(-4, 4, 500)
# Gauss
y = (1/(np.sqrt(2 * np.pi) * sigma)) * np.exp(-0.5 * (x)**2)

plt.plot(x, y, label='Gaussian laser pulse')
plt.fill_between(x, y, alpha=0.3)
plt.legend()

# Set custom x-ticks with sigma and my notation
xticks = np.arange(-4, 5, 1)
xtick_labels = [f"{i}σ" if i != 0 else 'μ' for i in xticks]
plt.xticks(xticks, xtick_labels)

# Calculate FWHM for Gaussian: FWHM = 2*sqrt(2*ln2)*sigma
fwhm = 2 * np.sqrt(2 * np.log(2)) * sigma
half_max = y.max() / 2

# FWHM limits in sigma units
left = -fwhm / 2
right = fwhm / 2

# Plot the FWHM line
plt.hlines(half_max, left, right, colors='black', linestyles='-', linewidth=2, label='FWHM')
plt.vlines([left, right], 0, half_max, colors='black', linestyles='--', linewidth=1)

# Notation FWHM
plt.annotate('FWHM', xy=(0, half_max), xytext=(0, half_max*1.2),
             ha='center', color='black', fontsize=10,
             arrowprops=dict(arrowstyle='-|>', color='black'))


# define the start of the pulse
peak_left = left - fwhm /2
peak_right = left + fwhm/2
peak_mid = (peak_left + peak_right)/2

# plot
plt.hlines(y.max(), peak_left, peak_right, colors='black', linestyles='-', linewidth=2, label='FWHM')
plt.vlines([peak_left], 0, y.max(), colors='black', linestyles='--', linewidth=1)

# Notation time until peak 
plt.annotate('Time until\npeak', xy=(peak_mid, y.max()), xytext=(peak_mid, y.max()-0.07),
             ha='center', color='black', fontsize=10,
             arrowprops=dict(arrowstyle='-|>', color='black'))

plt.grid(True)
plt.show()
```
