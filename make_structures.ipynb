```python
import os
import numpy as np
import subprocess
import re
```

```python
dirname = 'gro_structures'
efield = '3e4' #3e4 or 5e4
# Define the paths and create necessary directories
cwd = f'/home/mans/E_{efield}'
os.chdir(cwd)

#new directory for this simulation
os.makedirs(f'{dirname}', exist_ok=True)

E_folder = os.listdir(cwd) # Ã¤ndra denna till mapp med samma fold!
print(E_folder)
all_runs = [i for i in E_folder if re.match(r'run\\d+', i)]
all_runs.sort()
print(all_runs)
```

```python

t_end = 100000*0.0005
t_start = 0
num_sims = 101

moldstruct_path = "/home/spidocstester/MolDStruct/bin"  # Path to GROMACS executables (via MolDStruct)
trjconv_path = f"{moldstruct_path}/trjconv"  # Path to trjconv 

for run in all_runs:
    os.chdir(f'{cwd}/{dirname}')
    os.makedirs(f'{run}', exist_ok = True)
    newcwd = f'{cwd}/{dirname}/{run}'
    os.chdir(newcwd)
    trr_path = f'/home/mans/E_{efield}/{run}/long.trr'
    gro_path = f'/home/mans/E_{efield}/{run}/long.gro'
    
    for k,t in enumerate(np.linspace(t_start,t_end,num_sims)):
        output = f"{newcwd}/E_{efield}_{run}_{k:04d}.gro"
        # Run Tarjconv command to generate starting structures out of trajectory files from E-field simulations made by A.Sinelnikova
        trjconv_cmd = [trjconv_path, "-f", trr_path, "-s", gro_path, "-dump", str(t), "-o", output]
        print(trjconv_cmd)
        try:
            result = subprocess.run(trjconv_cmd, input='0 \n',check=True,text=True,capture_output=True)  # Join the command list into a single string
        except subprocess.CalledProcessError as e:
            print(f'Error in trjconv command: {e}')
            print(f'Error output: {e.stderr}')
```
