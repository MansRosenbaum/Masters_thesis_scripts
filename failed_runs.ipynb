```python
import os

```

```python
#For 5e4
efield = '3e4'
cwd = f'/home/mans/moldstruct_python/E_{efield}_mds'

list_fails = []

run_dirs = os.listdir(f'{cwd}')
run_dirs.sort()
run_dirs = run_dirs[1:] #Contains all the runs (runxxx_mds)

for run in run_dirs: # the one it is on
    
    fold_dirs = os.listdir(f'{cwd}/{run}') #Contains all the folds for that run (E_5e4_runxxx_0xxx_mds)
    fold_dirs.sort()    
    fold_dirs = fold_dirs[:-1]
    
    for fold in fold_dirs:
        try:
            content = (os.listdir(f'{cwd}/{run}/{fold}/'))
            content.sort()
            # Check if each folders' content is exaclty these files
            if content != [f'.{fold[:-4]}_exp.trr_offsets.npz', f'{fold[:-4]}.gro', f'{fold[:-4]}_exp.gro', f'{fold[:-4]}_exp.trr', 'additional_data', 'data.h5', 'ener.edr', 'kin.xvg', 'tot.xvg']:
                print(f'{fold}: {content}')
                list_fails.append(fold)
        except Exception as e:
            print(f'There is no {fold[:-4]}: {e}')

print('\nTotal number of fails: ', len(list_fails))
print(f'All fails are: {list_fails}')


'''Example: ['.E_5e4_run0_0000_exp.trr_offsets.npz', 'E_5e4_run0_0000.gro', 'E_5e4_run0_0000_exp.gro', 
'E_5e4_run0_0000_exp.trr', 'additional_data', 'data.h5', 'ener.edr', 'kin.xvg', 'tot.xvg']'''
```
